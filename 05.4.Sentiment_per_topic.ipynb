```python
import pandas as pd
from plotnine import ggplot, aes, geom_bar, labs,theme_light,theme,element_text,element_blank,scale_fill_manual,element_line,xlim,scale_x_continuous, scale_fill_gradient
```

```python
#Cargar el dataset

data = pd.read_csv('data/data_sentiment_analisis.csv')

def read_serie_topics_serializada(lista):
    serie_cargada = pd.read_pickle(lista+'.pkl')
    return serie_cargada
```

```python

"""
    NMF (Non Negative Matrix Factorization)
"""
topics_asig_nmf = read_serie_topics_serializada('asig_topics/topics_asig_nmf')


data['topics_nmf'] = topics_asig_nmf
data['topics_nmf'] = data.topics_nmf.apply(lambda x: x+1)

# Calcular la media del score por cada tpico
media_por_topico = data.groupby('topics_nmf')['score_analysis'].mean().reset_index()

# Crear el grfico de barras o de puntos
plot = (ggplot(media_por_topico, aes(x='topics_nmf', y='score_analysis', fill='score_analysis')) +
        geom_bar(stat='identity') + # Usar geom_bar para grfico de barras o geom_point para grfico de puntos
        scale_fill_gradient(low='red', high='green') +
        labs(title='Media del Score por Tpico NMF', x='Tpico', y='Media del Score')+
        theme_light()+
        theme(panel_grid_major=element_blank(),  # Quitar cuadrculas mayores
        panel_grid_minor=element_blank())
        + scale_x_continuous(breaks=list(range(0, 11))))
       
# Mostrar el grfico
plot.show()
```

```python
plot.save('image_topics/media_score_topics_nmf')
```

```python
"""
    LDA
"""

topics_asig_lda = read_serie_topics_serializada('asig_topics/topics_asig_lda')

data['topics_lda'] = topics_asig_lda
data['topics_lda'] = data.topics_lda.apply(lambda x: x+1)

# Calcular la media del score por cada tpico
media_por_topico = data.groupby('topics_lda')['score_analysis'].mean().reset_index()

# Crear el grfico de barras o de puntos
plot = (ggplot(media_por_topico, aes(x='topics_lda', y='score_analysis', fill='score_analysis')) +
        geom_bar(stat='identity') +  # Usar geom_bar para grfico de barras o geom_point para grfico de puntos
        scale_fill_gradient(low='red', high='green') +
        labs(title='Media del Score por Tpico LDA', x='Tpico', y='Media del Score')+
        theme_light()+
        theme(panel_grid_major=element_blank(),  # Quitar cuadrculas mayores
        panel_grid_minor=element_blank())
        + scale_x_continuous(breaks=list(range(0, 11))))
       

# Mostrar el grfico
plot.show()
```

```python
plot.save('image_topics/media_score_topics_lda')
```
